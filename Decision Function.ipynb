# This code generates two sets of data: one linearly separable and one non-linearly separable. It then trains a perceptron on each dataset, and plots the decision boundaries for both sets of data. Finally, it prints out the classification accuracy for each perceptron.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
from sklearn.linear_model import Perceptron

# Generate separable linear data
X1, y1 = make_blobs(n_samples=100, centers=[(-2,2), (2,-2)], cluster_std=0.5, random_state=42)
plt.scatter(X1[:,0], X1[:,1], c=y1)
plt.show()
#The code above imports necessary libraries such as numpy, matplotlib, and the Perceptron model from the scikit-learn library. Then it generates 100 random samples with two clusters that are separated linearly, using the make_blobs function from scikit-learn. The clusters are centered at (-2,2) and (2,-2) with a standard deviation of 0.5, and are stored in variables X1 and y1. Finally, the scatter plot of X1 with y1 as the color of the data points is displayed using the scatter function from matplotlib.


# Train perceptron on linearly separable data
clf = Perceptron(random_state=42)
clf.fit(X1, y1)
#The code above initializes a Perceptron object clf with random state 42 and trains it on the data stored in X1 and y1 using the fit method of the Perceptron object.

X2, y2 = make_moons(n_samples=100, noise=0.1, random_state=42)
plt.scatter(X2[:,0], X2[:,1], c=y2)
plt.show()
#The code above imports the make_moons function from scikit-learn, which generates two interleaving half circles, and then generates 100 samples of this type of data with 0.1 noise using a random state of 42, storing it in variables X2 and y2. Then, a scatter plot of X2 with y2 as the color of the data points is displayed using the scatter function from matplotlib.


# Train perceptron on non-separable linear data
clf2 = Perceptron(random_state=42)
clf2.fit(X2, y2)
#The code above initializes a new Perceptron object clf2 with the same random state of 42 as clf and trains it on the non-linearly separable data stored in X2 and y2 using the fit method of the Perceptron object.


# Plot decision boundaries
x_min, x_max = -5, 5
y_min, y_max = -5, 5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
Z1 = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
Z2 = clf2.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
fig, ax = plt.subplots(1,2,figsize=(10,5))
ax[0].contourf(xx, yy, Z1, alpha=0.3)
ax[0].scatter(X1[:,0], X1[:,1], c=y1)
ax[0].set_xlim(x_min, x_max)
ax[0].set_ylim(y_min, y_max)
ax[0].set_title('Linearly Separable Data')
ax[1].contourf(xx, yy, Z2, alpha=0.3)
ax[1].scatter(X2[:,0], X2[:,1], c=y2)
ax[1].set_xlim(x_min, x_max)
ax[1].set_ylim(y_min, y_max)
ax[1].set_title('Non-Linearly Separable Data')
plt.show()



# Print classification accuracy
print("Linearly separable data accuracy:", clf.score(X1, y1))
print("Non-linearly separable data accuracy:", clf2.score(X2, y2))
```
